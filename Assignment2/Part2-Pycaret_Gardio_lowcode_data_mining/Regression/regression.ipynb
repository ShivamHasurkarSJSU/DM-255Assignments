```python
from pycaret.regression import *
s = setup(data, target = 'Open', session_id = 123)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
pred_holdout = predict_model(best)
```

```python
new_data = data.copy().drop('Open', axis = 1)
predictions = predict_model(best, data = new_data)
```
