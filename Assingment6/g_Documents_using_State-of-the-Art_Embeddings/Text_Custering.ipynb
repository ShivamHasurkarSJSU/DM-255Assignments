```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
import numpy as np
```

```python
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
corpus = [
    "The quick brown fox jumps over the lazy dog.",
"Artificial intelligence has transformed many aspects of our lives.",
"Climate change is one of the most pressing issues of our time.",
"The stock market fluctuated wildly due to the unexpected news.",
"Advancements in technology have made remote work more feasible and productive.",
"Nutrition is crucial for maintaining good health and well-being.",
"The exploration of space has yielded a wealth of knowledge about our universe.",
"Political debates often reveal the diverse opinions within a society.",
"Investing in education is investing in the future of the country.",
"Music has the power to transcend cultural barriers and unite people.",
"Sustainable energy solutions are essential for our planet's future.",
"The impact of social media on communication is profound and far-reaching.",
"Mental health is just as important as physical health.",
"Innovation in the automotive industry is steering towards electric vehicles.",
"Preserving historical landmarks is important for cultural heritage.",
"The role of artificial intelligence in healthcare is expanding rapidly.",
"Understanding genetics is key to unlocking personalized medicine.",
"Global travel has become more accessible, connecting people and cultures.",
"The rise of e-commerce has changed the landscape of retail.",
"Cybersecurity is critical in protecting digital information.",
    ]
```

```python
corpus_embeddings = embedder.encode(corpus)
corpus_embeddings = corpus_embeddings /  np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
corpus_embeddings
```

```python
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5) #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_
```

```python
clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])
```

```python
for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    print(cluster)
    print("")
```
